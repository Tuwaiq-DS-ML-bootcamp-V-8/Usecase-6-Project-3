### Dealing with outliers:

```python
riyadh_aqqar_df.describe()
```

```python
# select numerical columns using pd_to_numeric
numric_df = riyadh_aqqar_df.select_dtypes(include=['int64', 'float64'])
zscore = numric_df.apply(zscore)
```

```python
abs(zscore)
```

```python
threshold = 3
outliers = riyadh_aqqar_df[(abs(zscore) > threshold).any(axis=1)]
```

```python
outliers
```

```python
# length before removing outliers
len(riyadh_aqqar_df)
```

```python
# drop rows that is equal to outliers
riyadh_aqqar_df = riyadh_aqqar_df[~riyadh_aqqar_df.isin(outliers)].dropna()
```

```python
# length after removing outliers
len(riyadh_aqqar_df)
```

#### Q1

```python
import plotly.express as px
# حساب متوسط سعر المتر لكل حي
expensive_hoods = riyadh_aqqar_df.groupby('الحي')['سعر المتر'].mean().reset_index()

# ترتيب الأحياء تصاعديًا حسب متوسط سعر المتر
df_sorted = expensive_hoods.sort_values('سعر المتر', ascending=False)

# اختيار أغلى 10 أحياء
top_10 = df_sorted.head(10)

# رسم مخطط شريطي باستخدام Plotly
fig = px.bar(top_10, x='الحي', y='سعر المتر', title='أغلى 10 أحياء في الرياض')
fig.show()
```

```python
expensive_hoods.to_csv('expensive_hoods.csv', index=False)
```

### Q2

```python
land_count_by_hood = riyadh_aqqar_df["الحي"].value_counts()
```

```python
land_count_by_hood_df = pd.DataFrame({
    'الحي': land_count_by_hood.index,
    'عدد الأراضي': land_count_by_hood.values
})
```

```python
# رسم مخطط شريطي باستخدام Plotly
fig = px.bar(land_count_by_hood_df.head(10), x='الحي', y='عدد الأراضي', title=' اكثر عدد اراضي معروضة للبيع في الرياض حسب الحي')
fig.show()
```

```python
land_count_by_hood_df.to_csv('land_count_by_hood.csv', index=False)
```

### Q3

```python
fig = px.scatter(riyadh_aqqar_df, x='المساحة', y='سعر المتر', color='الحي',
                  title='العلاقة بين المساحة وسعر المتر في أحياء الرياض')
fig.show()
```

```python
top_hoods = land_count_by_hood_df["الحي"].head(10).values.tolist()
top_hoods
```

### Q4

```python
neighborhood_purpose_counts = riyadh_aqqar_df.groupby(['الحي', 'الغرض']).size().reset_index(name='عدد العقارات')
neighborhood_purpose_counts = neighborhood_purpose_counts[neighborhood_purpose_counts["الحي"].isin(top_hoods)]

fig = px.bar(neighborhood_purpose_counts, x='الحي', y='عدد العقارات', color='الغرض', 
             title="عدد الاراضي في كل حي حسب الغرض", barmode='group')
fig.update_layout(xaxis_title="الحي", yaxis_title="العدد", xaxis_tickangle=-45)
fig.show()
```

```python
neighborhood_purpose_counts.to_csv('neighborhood_purpose_counts.csv', index=False)
```
